```python
import pandas as pd
import numpy as np
```

```python
with open('data.txt') as f:
    lines = f.readlines()
```

```python
for i in lines:
    print(i)
```

```python
Input = []
Output = []
```

```python
for i in lines:
    j = i.strip()
    Input.append(j)
```

```python
ques_phrase = [ 'which', 'what', 'whose', 'who', 'whom', 'whose', 'what', 'which', 'where', 'when', 'how', 'why' ]
```

```python
for sen in Input:
    if any(x in sen for x in ques_phrase):
        print(sen.split())
```

```python
for sen in Input:
    if any(x in sen.split() for x in ques_phrase):
        Output.append((str(sen) + ' Yes'))
    else:
        Output.append((str(sen) + ' No'))
```

```python
for b in Output:
    print(b)
```

```python
with open('data0.txt', 'w+') as g:
    for item in Output:
        g.write('{}\n'.format(item))
```

```python
g.close()
```
